```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# import matplotlib.pyplot as plt
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "Nimbus Sans"
})

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "Nimbus Sans",
    'font.size':24
})
```

```python
maizedata=pd.read_csv('../maizedata/maize_ActivityAnnotated.csv')
sorghumdata=pd.read_csv('../sorghumdata/Sorghum_ActivityAnnotated.csv')
```

```python
# maizedata
```

```python
maizedata['species']='maize'
sorghumdata['species']='sorghum'
```

```python
summary=pd.concat([maizedata, sorghumdata], axis=0)
summary=summary.groupby(["species", "RecategorizeAction"]).size().reset_index(name="count")
for _, row in summary.iterrows():
    if row['species']=='maize':
        summary.at[_, 'proportion']=row['count']/2005
    else:
        summary.at[_, 'proportion']=row['count']/1854

# summary
summary_df=summary.pivot(index='species', columns='RecategorizeAction', values='proportion')
```

```python

plt.figure(figsize=(5, 5))
# Set species as index for plotting
# plot_df = summary_df.set_index("species")[["Activator", "Repressor", "Dual Action", "No activity"]]

colors = {
    "Activator": "lightskyblue",
    "Repressor": "red",
    "Dual Action": "orange",
    "No Activity": "dimgray"
}

ax = summary_df.plot(
    kind="bar",
    stacked=True,
    figsize=(5,5),
    width=0.3,
    color=[colors[c] for c in summary_df.columns]
)

ax.legend(
    title="RecategorizeAction",
    bbox_to_anchor=(1.05, 0.5),  # move legend to the right
    loc="center left",
    frameon=False
)

plt.xticks(rotation=0)

plt.ylabel('Proportion of TFs tested')


plt.savefig('TFs_Stats_recategorize.png', dpi=300, bbox_inches='tight')
```

```python
summary=pd.concat([maizedata, sorghumdata], axis=0)
summary=summary.groupby(["species", "Action"]).size().reset_index(name="count")
for _, row in summary.iterrows():
    if row['species']=='maize':
        summary.at[_, 'proportion']=row['count']/2005
    else:
        summary.at[_, 'proportion']=row['count']/1854

# summary
summary_df=summary.pivot(index='species', columns='Action', values='proportion')

plt.figure(figsize=(5, 5))
# Set species as index for plotting
# plot_df = summary_df.set_index("species")[["Activator", "Repressor", "Dual Action", "No activity"]]

colors = {
    "Activator": "lightskyblue",
    "Repressor": "red",
    "Dual Action": "orange",
    "No activity": "dimgray"
}

ax = summary_df.plot(
    kind="bar",
    stacked=True,
    figsize=(5,5),
    width=0.3,
    color=[colors[c] for c in summary_df.columns]
)

ax.legend(
    title="Action",
    bbox_to_anchor=(1.05, 0.5),  # move legend to the right
    loc="center left",
    frameon=False
)

plt.xticks(rotation=0)

plt.ylabel('Proportion of TFs tested')


plt.savefig('TFs_Stats.png', dpi=300, bbox_inches='tight')
```
