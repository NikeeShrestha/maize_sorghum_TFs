```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "Nimbus Sans"
})

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "Nimbus Sans",
    'font.size':24
})
```

```python
# geneexpression=pd.read_csv('../maizedata/counts.NE2020.693.filtered_bcSP.orderSampler.ordered_coeff.txt', index_col=0)

geneexpression=pd.read_csv('../maizedata/counts.NE2020.693.filtered_bcSP.orderSampler.ordered_coeff.txt', index_col=0)
# geneexpression
```

```python
meanexpression=pd.DataFrame(geneexpression.mean())
# meanexpression['gene']=meanexpression.index

meanexpression=meanexpression.reset_index()

meanexpression.columns = ['gene_ID','expression']
# meanexpression
```

```python
TFs=pd.read_csv('../maizedata/maize_ActivityAnnotated.csv')

# TFs=pd.read_csv('../sorghumdata/Sorghum_ActivityAnnotated.csv')
# TFs
```

```python
mergedf=pd.merge(TFs, meanexpression, on='gene_ID')
# mergedf.shape
```

```python
sns.boxplot(data=mergedf, x='RecategorizeAction', y='expression')
plt.ylim(-10,50)
plt.ylabel('maize gene expression')

plt.savefig('../results/maize_recategorizedactivation_expression.png', dpi=300, bbox_inches='tight')
```

```python
sns.boxplot(data=mergedf, x='Action', y='expression')
# plt.ylim(-10,50)
plt.ylabel('maize gene expression')

plt.savefig('../results/maize_activation_expression.png', dpi=300, bbox_inches='tight')
```
